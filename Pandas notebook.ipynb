```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file = "HeroesOfPymoli/Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file)

#Confirm data has loaded 
purchase_data.head()
```

```python
#Player Count

#Display total number of players

total_players= len(purchase_data["SN"].unique())

summary_table=pd.DataFrame({"Total Players":[total_players]})

summary_table.head()
```

```python
#Purchasing Analysis (Total)

#Number of Unique Items
#Average Purchase price
#Total Number of Purchases
#Total Revenue 


unique_items= len(purchase_data["Item ID"].unique())

average_price=purchase_data["Price"].mean()

purchase_total=purchase_data["Price"].count()

total_revenue=purchase_data["Price"].sum()

summary_table=pd.DataFrame({"Number of Unique Items":[unique_items], "Average Purchase Price":[average_price], "Total Number of Purchases":[purchase_total], "Total Revenue":[total_revenue]})

summary_table=summary_table.round(2)

summary_table.head()

#try change to $
```

```python
#Gender Demographics 

# % and count of male players
# % and count of female players
# % and count of other/ non-disclosed

gender_count=purchase_data["Gender"].value_counts()

percent_gender=gender_count/total_players*100

gender_count

gender_demographics=pd.DataFrame({"Total Count":gender_count , "Percentage of Players": percent_gender})

gender_demographics=gender_demographics.round(2)

gender_demographics.head()

#cant get to show correct figures, keeps coutning all vales not unique 
```

```python
#Purchasing Analysis (Gender)

#Purchase count - by gender
#Average purchase price - by gender
#total purchase value - by gender
#average purchase total per person by gender

purchase_count=purchase_data.groupby(["Gender"]).count()["Price"]

average_purchase_price=purchase_data.groupby(["Gender"]).mean()["Price"]

total_purchase_value=purchase_data.groupby(["Gender"]).sum()["Price"]

total_purchase_analysis=total_purchase_value/gender_count

purchasing_analysis=pd.DataFrame({"Purchase Count": purchase_count , "Average Purchase Price": average_purchase_price , "Total Purchase Value": total_purchase_value , "AVG Total Perchase per Person": total_purchase_analysis})

purchasing_analysis=purchasing_analysis.round(2)

purchasing_analysis.head()
```

```python
#Age demographics

#into bins of 4 years
#purchase count
#average purchase price
#total purchase value
#average purchase total per person by age group

bins=[0,10,15,20,25,30,35,40,45]

group_names=["0-10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-40", "40+"]

pd.cut(purchase_data["SN"], bins, labels=group_names).head

```

```python
#Top Spenders

#top 5 spenders by total purchase value 
#sn
#purchase count
#average purchase price
#total purchase value 



#Purchase count
purchase_count=purchase_data['SN'].value_counts()

purchase_count


#Average purchase price

purchase_price = purchase_data.groupby("SN")

average_purchase = purchase_price.mean(["Price"])

average_purchase 

#total purchase value  

total_value= purchase_data .groupby(["SN"]).sum(["Price"])

total_value 



spenders_summary=pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": average_purchase , "Total Purchase Price": total_value})

spenders_summary.sort_values("Total Purchase Value",ascending=False).round(2).head()

spenders_summary
```
